```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.set(
    font_scale=1.5,
    style="whitegrid",
    rc={'figure.figsize':(15,7)}
        )
```

```python
path = 'https://worldtable.info/gosudarstvo/tablica-rozhdaemosti-po-godam-rossija.html'
```

```python
df = pd.read_html(path, header=0)[0]
df = df.rename(columns={'Год':'year', 'Количество родившихся чел.':'born'})
df.head()
```

```python
df.born = df.born.str.replace(' ', '')
df = df.astype({'year': 'int16', 'born': 'int32'})
```

```python
df.info()
```

```python
ax = sns.lineplot(x='year', y='born', data=df)
ax.set_title('Рождаемость по годам')
ax.set_xlabel('Годы')
ax.set_ylabel('Родилось, тыс.')
ax.set_yticklabels(labels=np.divide(ax.get_yticks(), 1000).astype(int));
```
